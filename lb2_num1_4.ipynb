#2.2.11
Напишите программу на Python для создания гистограммы из DataFrame.  

```
from matplotlib import pyplot as plt
import numpy as np

data =[[4, 2, 4, 2, 2],
[8, 3, 7, 6, 5],
[5, 4, 7, 6, 5],
[7, 2, 7, 8, 3],
[6, 2, 4, 2, 2]]
X= np.arange(0, 10, 2)

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
plt.xlim(1, 11, 2)
plt.ylim(0, 8)
ax.minorticks_on()
ax.grid(which= 'major', linewidth= 1.5)
ax.grid(which= 'minor', linestyle= '--')
ax.bar(X+1.70, data[0], color = 'b', width= 0.15)
ax.bar(X+1.85, data[1], color = 'g', width= 0.15)
ax.bar(X+2.00, data[2], color = 'r', width= 0.15)
ax.bar(X+2.15, data[3], color = 'c', width= 0.15)
ax.bar(X+2.30, data[4], color = 'purple', width= 0.15)
```

# 2.3.2
Напишите программ на Python для создания круговой диаграммы с названием популярности языков программирования.  
Пример данных:
Языки программирования: Java, Python, PHP, JavaScript, C #, C ++
Популярность: 22.2, 17.6, 8.8, 8, 7.7, 6.7


```
from matplotlib import pyplot as plt 
import numpy as np 

fig=plt.figure()
ax=fig.add_axes([0, 0, 1, 1])
ax.axis('equal')
langs=['Java', 'Python', 'PHP', 'JavaScript', 'C #', 'C ++']
explode = [0.2, 0, 0, 0, 0, 0]
popularity=[ 22.2, 17.6, 8.8, 8, 7.7, 6.7]
ax.pie(popularity, labels=langs, autopct='%1.2f%%', shadow=True, explode=explode, wedgeprops={ 'edgecolor':'k'})
plt.title('Popularity of progrmming languages')
plt.show()
```

# 2.4.1
Напишите программу на Python, чтобы нарисовать график рассеяния со случайным распределением по осям X и Y, нанесенный на график друг против друга.  

```
import matplotlib.pyplot as plt
from pylab import randn
X = randn(200)
Y = randn(200)
plt.scatter(X,Y, color='r', s = 10)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
